## Missing Value Analysis

The dataset contains missing values in several columns.

- Age certification has a large number of missing values as many older or regional titles do not have official age ratings.
- IMDb score and IMDb votes are missing for some titles, possibly due to unavailable ratings on IMDb.
- Seasons column contains missing values mainly for movies, as seasons are applicable only for TV shows.
- TMDB popularity and TMDB score are missing for some records due to incomplete metadata.

These missing values are common in entertainment datasets and do not indicate data quality issues.


```
titles['imdb_score'].describe()
```

Min and max values are far from the median, indicating possible outliers.

```
plt.boxplot(titles['imdb_score'].dropna())
plt.title("Boxplot of IMDb Score")
plt.show()
```

Points outside the whiskers confirm the presence of outliers.

Outliers were identified using descriptive statistics and boxplot visualization.
The distribution showed extreme values away from the median, confirming the presence of outliers.
Therefore, median was used for imputing missing values as it is robust to outliers.


```
# Handling missing values
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].median())
```

```
titles['imdb_votes'] = titles['imdb_votes'].fillna(titles['imdb_votes'].median())
```

```
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].median())
```

```
# there are missing values for seasons as movies don't have seasons
titles['seasons'] = titles['seasons'].fillna(0)
```

```
titles['age_certification'] = titles['age_certification'].fillna('Unknown')
```

```
# Univariate analysis -
plt.figure()
plt.hist(titles['imdb_score'], bins=20)
plt.title("Distribution of IMDb Scores")
plt.xlabel("IMDb Score")
plt.ylabel("Frequency")
plt.show()
```

Around 70% of titles have IMDb scores between 6 and 8, indicating that most content receives moderate audience approval.

```

titles['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

Movies form the majority of content in the dataset compared to TV shows.


```
# Genre distribution
titles['genres'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.show()
```

The chart shows the most common genres in the dataset.
Drama and Comedy are the most frequent genres, indicating that a
large portion of movies and TV shows belong to these categories.

```
plt.boxplot(titles['runtime'].dropna())
plt.title("Runtime Distribution")
plt.ylabel("Minutes")
plt.show()

```

Most titles have runtime between 80 and 120 minutes.
Some extreme values indicate the presence of outliers such as short films or long documentaries.


```
# Bivariate analysis
titles.groupby('type')['imdb_score'].mean()
```

```
sns.boxplot(x='type', y='imdb_score', data=titles)
plt.title("IMDb Score by Content Type")
plt.show()
```

Movies and TV shows show similar IMDb score distributions, although movies display slightly higher variation.
This suggests that content type alone does not strongly determine audience ratings.


```
merged = pd.merge(titles, credits, on='id', how='left')
merged
```

```
# Trend Analysis
yearly = titles.groupby('release_year').size()

plt.figure()
plt.plot(yearly.index, yearly.values)
plt.title("Content Released Over Years")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```

## Correlation Analysis

'''Correlation analysis helps understand the relationship between numerical variables.
It shows how strongly two variables are related to each other.
Values range between -1 and 1.'''

'''1 → strong positive relationship
   0 → no relationship
  -1 → strong negative relationship'''

corr = titles[['runtime','imdb_score','imdb_votes']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Numerical Features")
plt.show()
```

### Interpretation

The correlation matrix shows a moderate positive correlation between
IMDb votes and IMDb score, suggesting that titles with more audience
engagement tend to receive higher ratings.

Runtime has a very weak correlation with IMDb score, indicating that
the duration of a movie or show does not strongly influence audience ratings.

## Key Insights

- Movies form the majority of content compared to TV shows.
- IMDb scores are mostly concentrated between 6 and 8.
- Content production has increased significantly after 2000.
- Drama is most popular genre in dataset.
- Titles with higher IMDb votes tend to have higher ratings.
- Runtime shows weak correlation with IMDb score.
